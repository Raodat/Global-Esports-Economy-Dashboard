```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

columns = ["Rank", "Game", "Total Earnings", "Total Number of Players Competed"]

def get_game_details(row):
    cells = row.find_elements(By.TAG_NAME, "td")
    texts = [cell.text.strip() for cell in cells]
    
    # Remove texts after numbers in columns
    def clean_stat(text):
        return (
            text.replace("Player", "")
                .replace("Players", "")
                .replace("s", "")                   # Had to remove the "s" characters too since player and players are distinct texts
                .replace("\n", " ")
                .strip()
        )


    return {
        "Rank": texts[0], 
        "Game": texts[1], 
        "Total Earnings": texts[2],              
        "Total Number of Players Competed": clean_stat(texts[3])      
        
    }

def main():
    game_data = []

    driver = webdriver.Chrome()

    try:
        for page_id in range(1, 5):
            url = f"https://www.esportsearnings.com/countries/bd/games-top-100?page={page_id}"
            driver.get(url)

            WebDriverWait(driver, 5).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table tbody tr"))
            )

            rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

            for row in rows:
                game = get_game_details(row)
                if game:
                    game_data.append(game)

                if len(game_data) >= 100:
                    break

            if len(game_data) >= 100:
                break

            for row in rows[:10]:
                cells = row.find_elements(By.TAG_NAME, "td")
                print(len(cells), [c.text for c in cells])


    finally:
        driver.quit()

    df = pd.DataFrame(game_data[:100], columns=columns)
    df.to_csv("top_games_in_Bangladesh.csv", index=False)

    print(df.head())
    print(f"Saved {len(df)} games.")

if __name__ == "__main__":
    main()
```
